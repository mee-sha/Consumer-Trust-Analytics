# Feature Engineering

## Business Objective

The objective of this notebook is to create meaningful features from the cleaned customer review dataset.

These engineered features will improve business analysis, SQL reporting, and Power BI dashboard development.

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/processed/cleaned_edtech_reviews.csv")
```

```python
df.head()
```

# Feature 1

## Review Length Category

```python
df["review_length_category"] = pd.cut(
    df["review_length"],
    bins=[0,50,150,float("inf")],
    labels=["Short","Medium","Long"]
)

df[
    ["review_length","review_length_category"]
].head()
```

# Feature 2

## Customer Sentiment

```python
def sentiment(rating):

    if rating <= 2:
        return "Negative"

    elif rating == 3:
        return "Neutral"

    else:
        return "Positive"


df["sentiment"] = df["rating"].apply(sentiment)

df[
    ["rating","sentiment"]
].head()
```

# Feature 3

## Review Quarter

```python
quarter_map = {

    "January":"Q1",
    "February":"Q1",
    "March":"Q1",

    "April":"Q2",
    "May":"Q2",
    "June":"Q2",

    "July":"Q3",
    "August":"Q3",
    "September":"Q3",

    "October":"Q4",
    "November":"Q4",
    "December":"Q4"

}

df["review_quarter"] = (
    df["review_month"]
      .map(quarter_map)
)

df[
    ["review_month","review_quarter"]
].head()
```

# Feature 4

## Helpful Review Flag

```python
df["helpful_review"] = (
    df["thumbs_up"] >= 5
)

df[
    ["thumbs_up","helpful_review"]
].head()
```

# Feature 5

## Review Type

```python
def review_type(length):

    if length < 50:
        return "Quick Feedback"

    elif length < 150:
        return "Detailed Review"

    else:
        return "Comprehensive Feedback"


df["review_type"] = (
    df["review_length"]
      .apply(review_type)
)

df[
    ["review_length","review_type"]
].head()
```

```python
df.to_csv(
    "../data/featured/final_edtech_reviews.csv",
    index=False
)
```

# Feature Engineering Summary

The following engineered features were created:

- Review Length Category
- Customer Sentiment
- Review Quarter
- Helpful Review Flag
- Review Type
