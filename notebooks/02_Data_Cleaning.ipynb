# Data Cleaning

## Objective

The objective of this notebook is to clean the raw customer review dataset collected from multiple EdTech platforms.

Cleaning includes:
- Handling missing values
- Removing duplicate records
- Correcting data types
- Validating data quality
- Saving the cleaned dataset

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/edtech_reviews.csv")
```

```python
df.shape
```

```python
df.info()
```

## Step 1 : Handle Missing Values

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=["review"])
```

```python
df.isnull().sum()
```

## Step 2 : Remove Duplicate Records

```python
df.duplicated().sum()
```

## Step 3 : Convert Data Types

```python
df.dtypes
```

```python
df["review_date"] = pd.to_datetime(df["review_date"])
```

```python
df.dtypes
```

```python
df.to_csv(
    "../data/processed/cleaned_edtech_reviews.csv",
    index=False
)
```

## Cleaning Summary

The dataset has been successfully cleaned.

Actions performed:

- Removed rows with missing review text.
- Verified duplicate records.
- Converted review_date to datetime.
- Saved cleaned dataset for analysis.
